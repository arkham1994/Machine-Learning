```python
import pandas as pd
```

```python
# Reading the dataset

df = pd.read_excel('C:/Users/ishan/Downloads/Sample_Longlist_Data.xlsx')
df.head()
```

```python
df.columns
```

```python
# Dropping unncessary columns

df.drop(['Unnamed: 18', 'Unnamed: 19','Unnamed: 20'], axis = 1, inplace = True)
df.head()
```

```python
# Checking for missing values
df.isnull().sum()
```

```python
# Modifying feature 'sellerproductcount' 

df['sellerproductcount'] = df['sellerproductcount'].dropna().apply(lambda x: str(x).split()[-2] )

def get_int(string):   # This function removes ',' from the text
    if len(string)>3:
        return str(string).split(',')[0] + str(string).split(',')[1]
    else:
        return string
df['sellerproductcount'] = df['sellerproductcount'].dropna().apply(get_int )
```

```python
df['sellerproductcount']
```

```python
# Parsing out +ve ratings count and percent into new columns

df['sellerratings_+ve_percent'] = df['sellerratings'].dropna().apply(lambda x : str(x).split()[0][:-1])
df['sellerratings_+ve_count'] = df['sellerratings'].dropna().apply(lambda x : str(x).split("(")[1].split()[0])
```

```python
df.head()
```

```python
# Creating a function that will return a list of all emails in the text

import re

def email(text):
    emails = re.findall(r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+", text)
    return emails  
```

```python
df['seller_email'] = df['sellerdetails'].dropna().apply(email)
df.head()
```

```python
# Extracting country information into new column

df['business_country'] = df['businessaddress'].dropna().apply(lambda x: str(x)[-2:])
df['business_country']
```

```python
df['business_country'].value_counts()
```

```python
df2 = df[~(df['business_country'] == 'CN')]    # Selecting only those records which are not from China i.e. 'CN'
df2.shape
```

```python
# Selection only relevant features for further analysis

features = ['sellerlink','sellerproductcount','sellerratings_+ve_percent','sellerratings_+ve_count','Max % of negative seller ratings - last 30 days',
           'Max % of negative seller ratings - last 90 days','Max % of negative seller ratings - last 12 months',
           'Hero Product 1 #ratings','Hero Product 2 #ratings']
```

```python
df3 = df2[features].copy()
df3.head()
```

```python
# Changing columns into their appropriate data type

df3.dropna(inplace=True)   # Droppping missing values as they are important in deciding best sellers

df3['sellerproductcount'] = df3['sellerproductcount'].astype(int)
df3['sellerratings_+ve_percent'] = df3['sellerratings_+ve_percent'].astype(int)
df3['sellerratings_+ve_count'] = df3['sellerratings_+ve_count'].astype(int)
```

```python
# We are going to create one feature which will contain total ratings of both top1 and top2 products

df3['top1_2 sales'] = df3['Hero Product 1 #ratings'] + df3['Hero Product 2 #ratings']
df4 = df3.drop(['Hero Product 1 #ratings','Hero Product 2 #ratings'], axis = 1)
```

```python
df4
```

```python
# Creating a new feature that will contain the weighted average % negative ratings
# We are giving more weight/importance to recent ratings and low weight/importance to older ratings

df4['avg_-ve_ratings'] = (3 * df4['Max % of negative seller ratings - last 30 days'] + 2 * df4['Max % of negative seller ratings - last 90 days']
                             + df4['Max % of negative seller ratings - last 12 months']) / 6
df4.head()
```

```python
df5 = df4.drop(['Max % of negative seller ratings - last 30 days',
           'Max % of negative seller ratings - last 90 days','Max % of negative seller ratings - last 12 months'], axis  = 1)
df5.head()
```

```python
# Removing records which have less than 10 +ve ratings or less than 500 combined product ratings
df6 = df5[(df5['sellerratings_+ve_count'] > 10) & (df5['top1_2 sales'] > 500)]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df6[['sellerproductcount','top1_2 sales']] = scaler.fit_transform(df6[['sellerproductcount','top1_2 sales']])
df6.head()
```

We are using 4 features to decide which seller is better
1. Product Count(C)
2. Seller +ve % Ratings(SR)
3. Total ratings for top Product(R)
4. Average _ve ratings(NR)

Best Seller Rank = R * (100 - NR)/100 + C * SR/100

```python
df6['BSR'] = df6['top1_2 sales'] * ((100 - df6['avg_-ve_ratings'])/100) + df6['sellerproductcount'] * (df6['sellerratings_+ve_percent']/100)
```

```python
df6.sort_values(by = 'BSR', ascending = False).head()
```
